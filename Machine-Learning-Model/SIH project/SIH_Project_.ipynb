Step 1: Import of libraries

```python
# Library for file manipulation
import pandas as pd

#Importing train_test_split for spliting dataset to Training & Testing
from sklearn.model_selection import train_test_split

# Importing  Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
# Importing  Logistic Regression Model
from sklearn.linear_model import LogisticRegression
# Importing  Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
# Importing  Naive Bayes Model
from sklearn.naive_bayes import GaussianNB

#Importing Performance Metric(Accuracy)
from sklearn.metrics import accuracy_score
```

Step 3: Data cleaning

```python
print("Checking for missing values in each column:")
print(df.isnull().sum())
```

```python
# Remove columns that have any NaN value
df = df.dropna(axis=1)

# Check remaining columns after removal
print("Remaining columns after removing columns with NaN:")
print(df.isnull().sum())
```

Step 4: Data Preprocessing

```python
df.columns
```

```python
# Applying preprocessing to the original dataset
X = df.drop(columns=["IP_Address", "Label"])
y = df["Label"]# Target Variable
```

```python
# Visualize rows and column x
X.shape
```

```python
# Visualize rows and column y
y.shape
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
# Visualize training data
print("Rows and columns of training data X_train:", X_train.shape)
print("Rows and columns of training data y_train:", X_train.shape)
```

Step 5: Training Models

```python
#Models to be evaluated
models = [
    # Naive Bayes Model
    GaussianNB(),

    # Logistic regression model
    LogisticRegression(random_state=50, max_iter=1000),

    # Decision Tree Model
    DecisionTreeClassifier(random_state=42),

    # Random forest model
    RandomForestClassifier(n_estimators=100, random_state=42)
]
```

```python
# Evaluate each model
for i, model in enumerate(models):
    model.fit(X_train, y_train)
    test_accuracy = accuracy_score(y_test, model.predict(X_test))
    print(f"Model {i+1}: {type(model).__name__}")
    print(f"Testing Accuracy: {test_accuracy}")
    print("-----------------")
```
